__[문제 1: 첫 번째 사냥]__

마법사는 몬스터를 사냥하기 위해 마법을 사용합니다.

마법을 사용하려면 지능이 50 이상이어야 합니다.

마법사의 지능이 50 이상인지 확인하세요.

```python
wizard = {"strength": 10, "agility": 20, "intelligence": 80, "health": 50}
warrior = {"strength": 80, "agility": 30, "intelligence": 20, "health": 100}
rogue = {"strength": 40, "agility": 80, "intelligence": 30, "health": 60}
healer = {"strength": 20, "agility": 40, "intelligence": 70, "health": 70}

print(wizard["intelligence"] > 50);
```

<br>

__[문제 2: 두 번째 사냥]__

전사는 강력한 공격을 하기 위해 자신의 힘을 사용합니다.

전사의 힘이 60 이상이고 체력이 80 이상인지 확인하세요.

```python
print(warrior["strength"] > 60 and warrior["health"] > 80);
```

__[문제 3: 세 번째 사냥]__

도적은 민첩함을 이용해 몬스터를 회피하고 공격합니다.

도적의 민첩이 70 이상이거나 힘이 50 이상인지 확인하세요.

<br>

__[문제 4: 네 번째 사냥]__

힐러는 파티의 체력을 회복시키기 위해 마법을 사용합니다.

힐러의 지능이 60 이상이고, 체력이 50 이상인지 확인하세요.

__[문제 5: 레벨업]__

모든 캐릭터가 사냥을 성공적으로 마쳤습니다. 이제 경험치를 얻어 레벨업을 해야 합니다.


각 캐릭터의 경험치가 100 이상인지 print()를 이용해 확인해주세요.

<br>

경험치는 아래와 같습니다.

마법사: 120

전사: 150

도적: 110

힐러: 90



```python
exp = {"wizard": 120, "warrior": 150, "rogue": 110, "healer": 90};

for key,val in exp.items() :
    if val > 100 :
        print(f"{key}은 경험치가 100이상입니다.");
    else :
        print(f"{key}은 경험치가 100미만입니다.");
```
